```python
import pandas as pd
import numpy as np
```

```python
# table 2: worst case violations (compas)
results = pd.read_csv("compas_most_harm.csv")
compas = pd.read_csv('..\\data\\compas-scores-cleaned.csv')
compas1 = compas[compas.race == 'African-American']
compas2 = compas[compas.race != 'African-American']
```

```python
#compas
with open('table2.tex', 'w') as output:
    output.write("\\" + "begin{table} \n")
    output.write("\\centering  \n")
    output.write("\\begin{tabular}{c|cc||cc} \n")
    output.write('Variable & \multicolumn{2}{c}{Population} & \multicolumn{2}{c}{Worst-case Violation} \\ \n')
    output.write(' & African-American& Other & African-American& Other \\ \n')
    output.write("\\hline \n")
    output.write("\\hline \n")
    output.write("Prior Felonies & {} \pm {} & {}\pm {} & {}\pm {} & {} \pm {} \\ \n".format(round(compas1['priors_count'].mean(), 2),
                                                                          round(compas1['priors_count'].var() ** 0.5, 2),
                                                            round(compas2['priors_count'].mean(), 2),
                                                            round(compas2['priors_count'].var() ** 0.5, 2),
                                                            round(results['priors_count_1'].mean(), 2),
                                                            round(results['priors_count_1'].var() ** 0.5, 2),
                                                            round(results['priors_count_0'].mean(), 2),
                                                            round(results['priors_count_0'].var() ** 0.5, 2)))
    output.write("Charge Degree & {} \pm {} & {}\pm {} & {}\pm {} & {} \pm {}  \\ \n".format(round(compas1['charge_degree'].mean(), 2),
                                                                  round(compas1['charge_degree'].var() ** 0.5, 2),
                                                            round(compas2['charge_degree'].mean(), 2),
                                                            round(compas2['charge_degree'].var() ** 0.5, 2),
                                                            round(results['charge_degree_1'].mean(), 2),
                                                            round(results['charge_degree_1'].var() ** 0.5, 2),
                                                            round(results['charge_degree_0'].mean(), 2),
                                                            round(results['charge_degree_0'].var() ** 0.5, 2)))
    output.write("Juvenile Felonies & {} $\pm$ {} & {}$\pm$ {} & {}$\pm$ {} & {} $\pm$ {}  \\ \n".format(round(compas1['juv_fel_count'].mean(), 2),
                                                            round(compas1['juv_fel_count'].var() ** 0.5, 2),
                                                            round(compas2['juv_fel_count'].mean(), 2),
                                                            round(compas2['juv_fel_count'].var() ** 0.5, 2),
                                                            round(results['juv_fel_count_1'].mean(), 2),
                                                            round(results['juv_fel_count_1'].var() ** 0.5, 2),
                                                            round(results['juv_fel_count_0'].mean(), 2),
                                                            round(results['juv_fel_count_1'].var() ** 0.5, 2)))
    output.write("Juvenile Misdemeanor & {} $\pm$ {} & {}$\pm$ {} & {}$\pm$ {} & {} $\pm$ {} \\ \n".format(round(compas1['juv_misd_count'].mean(), 2),
                                                            round(compas1['juv_misd_count'].var() ** 0.5, 2),
                                                            round(compas2['juv_misd_count'].mean(), 2),
                                                            round(compas2['juv_misd_count'].var() ** 0.5, 2),
                                                            round(results['juv_misd_count_1'].mean(), 2),
                                                            round(results['juv_misd_count_1'].var() ** 0.5, 2),
                                                            round(results['juv_misd_count_0'].mean(), 2),
                                                            round(results['juv_misd_count_0'].var() ** 0.5, 2),))
    output.write("High Risk & {} $\pm$ {} & {}$\pm$ {} & {}$\pm$ {} & {} $\pm${} \\ \n".format(round(compas1['high_risk'].mean(), 2),
                                                            round(compas1['high_risk'].var() ** 0.5, 2),
                                                            round(compas2['high_risk'].mean(), 2),
                                                            round(compas2['high_risk'].var() ** 0.5, 2),
                                                            round(results['high_risk_1'].mean(), 2),
                                                            round(results['high_risk_1'].var() ** 0.5, 2),
                                                            round(results['high_risk_0'].mean(), 2),
                                                            round(results['high_risk_0'].var() ** 0.5, 2),))

    output.write("\\end{tabular} \n")
    output.write("\\caption{Area Under the Curve (AUC) across demographics groups. A logistic regression is used to estimate the probability of failing. "
            "Values in bold characters show AUC with probability higher than $0.1$ to be higher than the AUC obtained for White students (for race groups) or for female students (for gender groups) "
            "or for White female (for gender and race groups).}" + '\n')
    output.write("\\label{tab: 2}  \n")
    output.write("\\end{table} \n")
output.close()
```
